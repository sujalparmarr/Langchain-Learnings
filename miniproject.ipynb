```python
import json
from langchain.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain.output_parsers import CommaSeparatedListOutputParser

# --- INPUT ---
topic = input("Enter the Topic: ")

# --- PROMPTS ---
prompt_summary = ChatPromptTemplate.from_messages([
    ("system", "Summarize this {topic} into 3-4 sentences.")
])

prompt_posts = ChatPromptTemplate.from_messages([
    ("system", 
     "From this summary, create three different social media posts:\n"
     "1. Twitter style (short, catchy)\n"
     "2. LinkedIn style (professional)\n"
     "3. Instagram style (casual, friendly)\n"
     "Separate each post with a comma.\n\nSummary:\n{summary}")
])

prompt_hashtags = ChatPromptTemplate.from_messages([
    ("system", "Generate 3 hashtags for this post:\n\n{post}")
])

# --- MODEL & PARSERS ---
model = ChatOpenAI(model='gpt-5-nano')
parser_str = StrOutputParser()
parser_list = CommaSeparatedListOutputParser()

summary = (prompt_summary | model | parser_str).invoke({"topic": topic})

posts = (prompt_posts | model | parser_list).invoke({"summary": summary})

final_posts = []
for post in posts:
    hashtags = (prompt_hashtags | model | parser_list).invoke({"post": post})
    final_posts.append({
        "post": post.strip(),
        "hashtags": hashtags
    })

# FINAL JSON STRUCTURE
final_output = {
    "topic": topic,
    "summary": summary,
    "posts": final_posts
}

# --- PRINT & SAVE ---
print(json.dumps(final_output, indent=2))

with open("ai_content_factory.json", "w", encoding="utf-8") as f:
    json.dump(final_output, f, indent=2, ensure_ascii=False)

print("\n✅ JSON saved to ai_content_factory.json")
```
